```python
import random
name = input("enter your name and surname")
age = int(input("Enter your age: "))
age_in_5_years = age + 5
print(f"in 5 years you will be {age_in_5_years} years old")
cat_names_input = input("enter cat names separated by commas: ")
cat_names = [name.strip() for name in cat_names_input.split(",") if name.strip()]
num_cats = len(cat_names)
print(f"You provided {num_cats} cat name(s)")
if num_cats >=2:
    random_cats = random.sample(cat_names, 2)
    concatenated_names = random_cats[0] + random_cats[1]
    print(f"Here's a concatenation of two random cat names: {concatenated_names}")
else:
      print("you need to provide at least 2 cat names for concatenation")
```
